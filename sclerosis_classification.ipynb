```python
# importing pandas module
import pandas as pd 
```

```python
# reading and creating a data frame of csv file
file = pd.read_csv('D:\\datasets\\sclerosis\\conversion_predictors_of_clinically_isolated_syndrome_to_multiple_sclerosis.csv',index_col=0)
df = file.copy()
#get 10 samples randomly
df.sample(10)
```

```python
#getting information from data frame
df.info()
```

```python
#how many nulls does data frame have ?
df.isnull().sum()
```

```python
#storing columns that have null data in a list
nan_cols = [col for col in df.columns if df[col].isnull().any()]
#value counting of data frame
for i in nan_cols :
    print(i)
    print(df[i].value_counts(ascending=False))
```

```python
#Initial_EDSS and Final_EDSS have many nulls so we should drop them because we cant impute them with mean or median and etc
df = df.drop(nan_cols[2:],axis=1)
df.info()
```

```python
#importing SimpleImputer and numpy modules
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values=np.nan)
```

```python
#impute data frame with mean strategy and make a data frame of that 
imputed_df = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
imputed_df.isnull().sum()
```

```python
#seperating label column and drop it from data frame
label = imputed_df['group']
imputed_df = imputed_df.drop(['group'],axis=1)
```

```python
imputed_df.describe()
```

```python
#seperating coulmns with binary values
binary_columns = list()
for col in imputed_df.columns :
    val = list(set(imputed_df[col].to_list()))
    if len(val) == 2 :
        if 1 in val and 0 in val:
            binary_columns.append(col)
binary_columns
```

```python
binary_df = imputed_df[binary_columns]
non_binary_df = imputed_df.drop(binary_df.columns,axis=1)
```

```python
#importing StandardScaler from sklearn module
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#scaling that data frame with non_binary columns
scaled_non_binary_df = pd.DataFrame(scaler.fit_transform(non_binary_df),columns=non_binary_df.columns)
scaled_non_binary_df.describe()
```

```python
#joining scaled data frame with data frame with binray values
finall_df = pd.concat([binary_df,scaled_non_binary_df],axis=1)
finall_df.head()
```

```python
#splitting our data to train and test 
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(finall_df,label,test_size=0.18,random_state=0)
```

```python
#making a model of Decision Tree with max_leaf_nodesm : 35
from sklearn.tree import DecisionTreeClassifier
max_leaf = 35
model = DecisionTreeClassifier(max_leaf_nodes=max_leaf)
```

```python
#fitting model with training data
model.fit(xtr,ytr)
```

```python
#testing with train data
model.score(xtr,ytr)*100
```

```python
#testing with test data
model.score(xts,yts)*100
```

```python
#testing with original data data
model.score(finall_df,label)*100
```

```python
#testing accuracy of model
bad_result = 0
for i in range(20) :
    xtr , xts , ytr , yts = train_test_split(finall_df,label,test_size=0.18,random_state=0)
    model = DecisionTreeClassifier(max_leaf_nodes = max_leaf)
    model.fit(xtr,ytr)
    res = (model.score(xtr,ytr)*100) - (model.score(xts,yts)*100)
    if res > 15 :
        bad_result += 1
bad_result
```
